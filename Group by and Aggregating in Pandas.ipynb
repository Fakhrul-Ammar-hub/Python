# Group by and Aggregating


```python
#Group by
```

```python
import pandas as pd
```

```python
df = pd.read_csv('D:\Document\Data Analyst Bootcamp\Python\csv files\companies.csv')
df
```

```python
group_by_frame = df.groupby('Industry')
```

```python
group_by_frame.count()
```

```python
group_by_frame.mean()
```

```python
group_by_frame.min()
```

```python
group_by_frame.max()
```

```python
group_by_frame.sum()
```

```python
group_by_frame.agg({'Revenue(billions US$)': ['mean', 'max','count','sum']})
```

```python
group_by_frame.agg({'Revenue(billions US$)': ['mean', 'max','count','sum'],'Profit(billions US$)':['mean', 'max','count','sum']})
```
